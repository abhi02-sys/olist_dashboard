##Finding insights for completed orders

```python
#creating completed orders df from Orders that were completed
df_completed = df[df['order_status']=='delivered']
```

```python
df_completed['order_status'].value_counts()
```

```python
df_completed.head()
```

Total Orders and Revenue

```python
total_orders = df_completed['order_id'].nunique()
print("Total Orders:" ,total_orders)
total_revenue = df_completed['total_payment_value'].sum()
print("Total Revenue:", round(total_revenue,2))
avg_order_value = total_revenue/ total_orders
print("Average Order Value:" , round(avg_order_value,2))
```

Time series Analysis to find about year, month of orders

```python
df_completed['order_year']= df_completed['order_purchase_timestamp'].dt.year
df_completed['order_month'] = df_completed['order_purchase_timestamp'].dt.month
```

```python
df_completed["order_year"]
df_completed['order_month']
```

```python
df_completed.head()
```

```python
#Year wise orders completed
df_completed.groupby('order_year')['order_id'].nunique()
```

```python
df_completed["order_year_month"] = df_completed["order_purchase_timestamp"].dt.to_period("M")

monthly_trend = (
    df_completed.groupby("order_year_month")["order_id"]
    .nunique()
    .reset_index()
)

```

```python
monthly_trend
```

Revenue on orders yearly and year-month wise

```python
yearly_revenue = orders_trend = (
    df_completed.groupby("order_year")
    .agg(
        total_orders=("order_id", "nunique"),
        total_revenue=("total_payment_value", "sum")
    )
    .reset_index()
)
```

```python
yearly_revenue
```

```python
yr_month_revenue = df_completed.groupby('order_year_month').agg(
    total_orders = ('order_id','nunique'),
    total_revenue=('total_payment_value','sum')
).reset_index()
```

```python
yr_month_revenue
```

```python
aov_trend = (
    df_completed
    .groupby("order_year_month")
    .agg(
        total_revenue=("total_payment_value", "sum"),
        total_orders=("order_id", "nunique")
    )
    .reset_index()
)

aov_trend["AOV"] = aov_trend["total_revenue"] / aov_trend["total_orders"]
aov_trend
```

Delivery Performance

```python
df_completed["delivery_days"] = (
    df_completed["order_delivered_customer_date"] -
    df_completed["order_purchase_timestamp"]
).dt.days

df_completed["delivery_delay_days"] = (
    df_completed["order_delivered_customer_date"] -
    df_completed["order_estimated_delivery_date"]
).dt.days
```

```python
avg_delivery_days = df_completed["delivery_days"].mean()
late_delivery_pct = (df_completed["delivery_delay_days"] > 0).mean() * 100
print(avg_delivery_days)
print(late_delivery_pct)
```

state wise revenue and orders

```python
state_performance = (
    df_completed.groupby("customer_state")
    .agg(
        total_orders=("order_id", "nunique"),
        total_revenue=("total_payment_value", "sum"),
        avg_delivery_days=("delivery_days", "mean")
    )
    .reset_index()
)
```

```python
state_performance
```

customer analysis

```python
customer_orders = (
    df_completed.groupby("customer_unique_id")
    .agg(order_count=("order_id", "nunique"))
    .reset_index()
)

repeat_customer_pct = (customer_orders["order_count"] > 1).mean() * 100
```

```python
repeat_customer_pct
```

```python
repeat_customer_pct = round(float(repeat_customer_pct), 2)
print(f"Repeat Customer %: {repeat_customer_pct}%")
```

```python
#Revenue contribution by top states
state_performance.sort_values("total_revenue", ascending=False).head(5)
```

```python
#Creating csv's for Power BI
yearly_revenue.to_csv("yearly_revenue.csv", index=False)
yr_month_revenue.to_csv("monthly_revenue.csv", index=False)
aov_trend.to_csv("aov_trend.csv", index=False)
state_performance.to_csv("state_performance.csv", index=False)

# Optional KPI summary
kpi_summary = pd.DataFrame({
    "Total Orders": [total_orders],
    "Total Revenue": [total_revenue],
    "Average Order Value": [avg_order_value],
    "Repeat Customer %": [repeat_customer_pct],
    "Avg Delivery Days": [avg_delivery_days],
    "Late Delivery %": [late_delivery_pct]
})

kpi_summary.to_csv("kpi_summary.csv", index=False)
```
